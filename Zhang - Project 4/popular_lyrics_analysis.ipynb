```python
# import dependencies
from pathlib import Path
import pandas as pd
import io
# from tokenizers import ByteLevelBPETokenizer
```

```python
# import popular_lyrics CSV file
from google.colab import files


uploaded = files.upload()
```

```python
lyrics_df = pd.read_csv(io.BytesIO(uploaded['popular_lyrics.csv']))
lyrics_df.head()
```

```python
# create new df to house lyrics analysis
lyrics_sentiment_df = pd.DataFrame(columns=['track_name',
                                            'track_popularity',
                                            'album',
                                            'artist_name',
                                            'artist_genres',
                                            'artist_popularity',
                                            'sentiment'])
lyrics_sentiment_df
```

```python
from transformers import pipeline
```

```python
classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)

for i in range(3):
  text = lyrics_df['lyrics'][i]
  candidate_labels = ["happy", "sad", "neutral"]
  result = classifier(text, candidate_labels=candidate_labels)
  #print(result)
  new_row = pd.Series({'track_name': lyrics_df['track_name'][i],
                       'track_popularity': lyrics_df['track_popularity'][i],
                       'album': lyrics_df['album'][i],
                       'artist_name': lyrics_df['artist_name'][i],
                       'artist_genres': lyrics_df['artist_genres'][i],
                       'artist_popularity': lyrics_df['artist popularity'][i],
                       'sentiment': [print(result)]})
  lyrics_sentiment_df = pd.concat([lyrics_sentiment_df, new_row.to_frame().T], ignore_index=True)

lyrics_sentiment_df
```
