```python
import pandas as pd
from pathlib import Path
```

```python
music_df_load = Path("Resources/playlist_2010to2023.csv")
```

```python
music_df = pd.read_csv(music_df_load, encoding='ISO-8859-1') 
```

```python
music_df.head()
```

```python
music_df.shape
```

```python
music_df.drop(columns=['playlist_url', 'track_id', 'artist_id', 'mode','time_signature'], inplace=True)
```

```python
music_df.head()
```

```python
music_df['duration_ms'] = music_df['duration_ms'] / 60000
```

```python
music_df.head()
```

```python
music_df['duration_ms'] = music_df['duration_ms'].round(2)
```

```python
music_df.head()
```

```python
music_df = music_df.rename(columns={'duration_ms': 'duration_min'})
```
